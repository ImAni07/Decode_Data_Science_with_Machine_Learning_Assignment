Question 1.6:

Write a program to display the appropriate message as per the color of signal (RED-Stop/Yellow-Stay/ Green-Go) at the road crossing.

```python
# Question 1.6:

"""
    Program to Display Appropriate Message according to the Color of the Signal at the Road Crossing.
    The Signal - Message Pair is given below:
        Red - Stop
        Yellow - Stay
        Green - Go
"""

def traffic_signal (color):
    if color.casefold () == "red":
        return "Stop"
    elif color.casefold () == "yellow":
        return "Stay"
    elif color.casefold () == "green":
        return "Go"
    else:
        return "Invalid Color.\nEnter a Valid Color!!!"

signal_color = input("Enter the color of the traffic signal: ")
message = traffic_signal (signal_color)
print (f"The Color of the Signal is: ", signal_color)
print(f"Message: ", message)
```

Answer 1.6:

The objective of the above code is to Display Appropriate Message according to the Color of the Signal.

The explanation for the same is given below:
    i. The function traffic_signal (color) takes a color as input.

    ii. By using casefold() on the color variable, we ensure that the comparison is case-insensitive, so the program will work correctly regardless of whether the user enters the color in uppercase, lowercase, or a mix of both.
    
    iii. It checks the color of the signal and returns an appropriate message -
        If the color is "Red", it returns "Stop";
        if the color is "Yellow", it returns "Stay";
        If the color is "Green", it returns "Go";
        For any other Input, it returns "Invalid Color.\nEnter a Valid Color!!!"
    
    iv. The code prompts the user to input the color of the traffic signal and then prints both the Color of the Signal and the corresponding message.
    
    v. The Output of the code, for different scenarios are as follows -
        signal_color = "Red" : Output = "The Color of the Signal is: Red\nMessage: Stop"
        signal_color = "Yellow" : Output = "The color of the Signal is: Yellow\nMessage: Stay"
        signal_color = "Green" : Output = "The color of the Signal is: Green\nMessage: Go"
        For any other Input value of signal_color : Output = "Invalid Color.\nEnter a Valid Color!!!"

```python
# Question 1.8:

"""
    Program to find the larger of the three numbers
    Numbers are pre - specified
    Ternary Operators
"""

# Pre - Specified Numbers
n1 = 7
n2 = 18
n3 = 45

# Performing the Required Operation
result = n1 if (n1 > n2 and n1 > n3) else (n2 if n2 > n3 else n3)

# Display the Result
print (f"The Largest among {n1}, {n2} and {n3} is {result}")
```

Write a program to find the factors of a whole number  using a while loop.

```python
# Question 1.9:

# Program to find the Factors of a Whole Number using Whole Number

# Function to find factors of a number using a while loop
def find_factors(number):
    factors = []
    i = 1
    while i <= number:
        if number % i == 0:
            factors. append (i)
        i += 1
    return factors

# Take User Input
number = int (input ("Enter a whole number: "))

# Print the Output, based on Condition
if number > 0:
    factors = find_factors (number)
    print(f"The factors of {number} are: {factors}")
else:
    print("Please enter a positive whole number.")
```

```python
# Question 1.10:

"""
    Program to find the Sum of all Positive Numbers
    Take Input from User
    Negative Number to STOP the Program
"""

# Function to Add the Positive Numbers
def sum_positive_numbers():
    total_sum = 0
    num = []
    
    # Exception Handling, in case of any Invalid Input
    while True:
        try:
            n = float(input("Enter a Number: "))
            num. append (n)
            if n < 0:
                break
            total_sum += n
        except ValueError:
            print("Please enter a valid number.")
    
    # Print the Output
    print (f"The Numbers Entered by the User are:\n{num}\n")
    print (f"The sum of all positive numbers entered is: {total_sum}")

# Call the function to start the program
sum_positive_numbers ()
```

Question 1.11.

Write a program to find prime numbers between 2 to 100 using nested for loops.

```python
# Question 1.11:

# Program to find Prime Numbers between 2 to 100, using nested for loops

# Import Module
import math

# Initialize an empty list to store prime numbers
prime_num_list = []

# Iterate over numbers from 2 to 100
for n in range(2, 101):
    is_prime = True

    # Check if the number is divisible by any number up to its square root
    for i in range(2, int (math. sqrt (n)) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_num_list.append(n)

# Print the Output
print(f"Prime numbers between 2 and 100:\n{prime_num_list}\n")
print (f"There are {(len (prime_num_list))} Prime Numbers within the Specified Range of 2 and 100.")
```

Answer 1.11:

The above code is designed to find and print all prime numbers between 2 and 100.

Here's a breakdown of how it works:

    ii. Main Loop: The code iterates over numbers from 2 to 100 using a for loop with range(2, 101).
    
    iii. Checking for Primality:
        a. For each number 'n' in the range, a flag 'is_prime' is set to "True", assuming it's prime initially.
        b. The code then checks if 'n' is divisible by any number 'i' between 2 and the square root of 'n' (inclusive) using another for loop. For finding the square root of 'n', math module has been used.
        c. If 'n' is divisible by any 'i' (i.e., 'n % i == 0'), the 'is_prime' flag is set to "False" and the inner loop is exited using the 'break' statement.
        d. If the 'is_prime' flag remains "True" after the inner loop, it means 'n' is a prime number, and it's appended to the "prime_num_list".
        e. The code repeats this process for all numbers between 2 and 100, and the resulting list of prime numbers is printed at the end.

Write the  programs for the following:

a. Accept the marks of the student in five major subjects and display the same.

b. Calculate the sum of the marks of all subjects.Divide the total marks by number of subjects (i.e. 5), calculate percentage = total marks/5 and display the percentage.

c. Find the grade of the student as per the following criteria.
Hint: Use Match & case for this.

Criteria                          Grade
Percentage > 85                     A
Percentage < 85 and >= 75           B
Percentage < 75 and >= 50           C
Percentage > 30 and <= 50           D
Percentage < 30                  Reappear

```python
# Question 1.12:

"""
    Program to take user input for marks in 5 subjects
    Calculate Total Marks and Percentage
    Find the Grade
"""
# Function to accept subjects and display them
def accept_subjects ():
    subjects = []
    for i in range (5):
        subject = input (f"Enter subject {i+1} name: ")
        subjects. append (subject)
    return subjects

# Function to accept marks of 5 subjects and display them
def accept_marks (subjects):
    marks = {}
    for subject in subjects:
        mark = float (input (f"Enter marks for {subject}: "))
        marks [subject] = mark
    return marks

# Function to calculate total marks
def calculate_total_marks (marks):
    total_marks = sum (marks. values ())
    return total_marks

# Function to calculate percentage
def calculate_percentage (total_marks):
    percentage = total_marks / 5
    return percentage

# Function to determine the grade based on percentage
def determine_grade (percentage):
    match percentage:
        case p if p > 85:
            grade = 'A'
        case p if p >= 75:
            grade = 'B'
        case p if p >= 50:
            grade = 'C'
        case p if p > 30:
            grade = 'D'
        case _:
            grade = 'Reappear'
    return grade

# Main function to run the program
def main ():
    subjects = accept_subjects ()
    print(f"Subjects: {subjects}\n")
    
    marks = accept_marks (subjects)
    print("Marks entered:")
    for subject, mark in marks. items ():
        print(f"{subject}: {mark}")
    
    total_marks = calculate_total_marks (marks)
    print (f"\nTotal Marks: {total_marks}")
    
    percentage = calculate_percentage (total_marks)
    print(f"\nPercentage: {percentage}%")
    
    grade = determine_grade (percentage)
    print (f"\nGrade: {grade}")

if __name__ == "__main__":
    main ()
```

Question 1.13:

Write a program for VIBGYOR Spectrum based on their Wavelength using Wavelength Range:

Color                       Wavelength (nm)
Violet                      400.0 - 440.0
Indigo                      440.0 - 460.0
Blue                        460.0 - 500.0
Green                       500.0 - 570.0
Yellow                      570.0 - 590.0
Orange                      590.0 - 620.0
Red                         620.0 - 720.0

```python
# Question 1.13:

# Program to Determine the Color based on their Wavelength (VIBGYOR Spectrum)

# Function to Determine the Color
def find_color (wavelength):
    if 400.0 <= wavelength < 440.0:
        return "Violet"
    elif 440.0 <= wavelength < 460.0:
        return "Indigo"
    elif 460.0 <= wavelength < 500.0:
        return "Blue"
    elif 500.0 <= wavelength < 570.0:
        return "Green"
    elif 570.0 <= wavelength < 590.0:
        return "Yellow"
    elif 590 <= wavelength < 620.0:
        return "Orange"
    elif 620.0 <= wavelength <= 720.0:
        return "Red"
    else:
        return "Wavelength out of Visible Range!!!\nPlease Enter a Valid Wavelenth."

# Main Function
def main ():
    
    # Exception Handling
    try:
        wavelength = float (input ("Enter the Wavelength (in nm): "))
        color = find_color (wavelength)
        print (f"\nThe Color corresponding to {wavelength} nm is {color}")
    except ValueError:
        print ("Please Enter a Valid Numeric Value for Wavelength!!!")

if __name__ == "__main__":
    main ()
```

i. Objective: This program is designed to take user input for wavelength of light in nm (nanometer), and then determine the color name corresponding to this wavelength.

ii. 'find_color (wavelength)' function:

    b. It uses a series of if-elif statements to check which range the wavelength falls into and returns the corresponding color.
    c. If the wavelength does not fall into any of the specified ranges, it returns "Wavelength out of visible range".

iii. Input the Wavelength:
    a. The program prompts the user to enter the wavelength in nanometers (nm).
    b. It tries to convert the input to a float. If the input is not a valid number, it catches the ValueError and prints an error message.

iv. 'if __name__ == "__main__"' function:

    b. In this block, the main function is called to execute the code. Therefore, whenthis script is executed, it will prompt to enter the wavelength and then print the corresponding color in the VINGYOR Spectrum.

Question 1.14:

Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system.

Given:

mass_earth = 5.972e24  # Mass of Earth in kilograms
mass_moon = 7.34767309e22  # Mass of Moon in kilograms
mass_sun = 1.989e30  # Mass of Sun in kilograms
distance_earth_sun = 1.496e11  # Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8  # Average distance between Moon and Earth in meters

Tasks:

    a. Calculate the gravitational force between the Earth and the Sun
    b. Calculate the gravitational force between the Moon and the Earth
    c. Compare the calculated forces to determine which gravitational force is stronger
    d. Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison

```python
# Question 1.14:

"""
    Program to calculate the Gravitational Force between Earth and Sun, as well as Moon and Earth
    Compare the calculated forces to determine which Gravitational Force is Stronger
    Explain which celestial body (Earth or Moon) is attracted to the other, based on the comparison
"""

# Import Module
import math

# Given Values
mass_earth = 5.972e24  # Mass of Earth in kilograms
mass_moon = 7.34767309e22  # Mass of Moon in kilograms
mass_sun = 1.989e30  # Mass of Sun in kilograms
distance_earth_sun = 1.496e11  # Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8  # Average distance between Moon and Earth in meters

# Constant Value, but not given
G = 6.674e-11  # Gravitational constant in N(m^2)/kg^2

# Calculations

# Gravitational force between the Earth and the Sun
force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun**2)

# Gravitational force between the Moon and the Earth
force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth**2)

# Results
print ("The gravitational force between the Earth and the Sun is {:.2e} N". format (force_earth_sun))
print ("The gravitational force between the Moon and the Earth is {:.2e} N". format (force_moon_earth))

# Conclusions

# Compare the calculated forces
if force_earth_sun > force_moon_earth:
    print ("The gravitational force between the Earth and the Sun is stronger.")
else:
    print ("The gravitational force between the Moon and the Earth is stronger.")

# Determine which celestial body is more attracted to the other
if force_earth_sun > force_moon_earth:
    print ("The Earth is more attracted to the Sun than the Moon is to the Earth.")
else:
    print ("The Moon is more attracted to the Earth than the Earth is to the Sun.")
```

ii. Initialization:
    a. First, the 'math' module is imported (the only necessary module)

iii. Calculation: The gravitational forces between the Earth and the Sun, and the Moon and the Earth, are calculated, using the formula:
                    "F = G * (m1 * m2) / r^2"
                    where,
                        F is the gravitational force,
                        G is the gravitational constant,
                        m1 and m2 are the masses of the two objects, and
                        r is the distance between them.

iv. Printing: The calculated forces are then printed, and the code compares the two forces to determine which is stronger. If the force between the Earth and the Sun is greater, the code prints that the Earth is more attracted to the Sun than the Moon is to the Earth. If the force between the Moon and the Earth is greater, the code prints that the Moon is more attracted to the Earth than the Earth is to the Sun.
